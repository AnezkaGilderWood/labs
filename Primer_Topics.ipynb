## Interpolation

In numerical approaches, we often evaluate a function (or a solution to an ODE) on a grid of points. For example, consider the following numerical approximation to $f(x) = \sin(2\pi x)$

```python
x = np.linspace(0, 1, 10)
f = np.sin(2 * np.pi * x)
fig, ax = plt.subplots()
ax.plot(x, f, "-x")
ax.set(xlabel="x", ylabel="f(x)")
plt.show()
```

But, what if we want to evaluate the function between two of the points on the grid? In the case above, you may just say "well, just evaluate the function at that point..". But, oftentimes it is not so simple, perhaps we obtained `f` from solving an ODE numerically and it would be very expensive to re-run out integration. This is where **interpolation** comes in to play. We can use standard routines from `scipy` to help us:

```python
from scipy.interpolate import interp1d

x_smooth = np.linspace(0, 1, 1000)
f_interpolated = interp1d(x, f)

fig, ax = plt.subplots()
ax.plot(x, f, "-x", label="exact values")
ax.plot(x_smooth, f_interpolated(x_smooth), "-", label="interpolated values")
ax.set(xlabel="x", ylabel="f(x)")
plt.show()
```
